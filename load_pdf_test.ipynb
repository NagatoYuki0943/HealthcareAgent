```python
pdf_path = "./data/FM docs 2024.3/Eye Pressure Lowering Effect of Vitamin C.pdf"
```

# pypdf

```python
from pypdf import PdfReader
```

```python
reader = PdfReader(pdf_path)
reader
```

```python
len(reader.pages)
```

```python
page = reader.pages[0]
page
```

```python
print(page.extract_text())
```

# pdfminer.six

```python
from pdfminer.high_level import extract_text
```

```python
print(extract_text(pdf_path))
```

# PyMuPDF

```python
import pymupdf
```

```python
doc = pymupdf.open(pdf_path)
doc
```

```python
len(doc)
```

```python
page = doc[0]
page
```

```python
print(page.get_text()) # get plain text encoded as UTF-8
```
